How does the size of the mitochondria in a cell depend on the size of the cell itself?

```r
install.packages("dplyr")
```

```r
# plot settings
options(repr.plot.width=10, repr.plot.height=7)
library(ggplot2)
library(reshape2)
library(dplyr)
# Simulate power-law data

N <- 30
alpha <- 2.3
Y0 <- 100
sigma <- 500

CellSize <- runif(N,3,10) # independent variable
eps <- rnorm(N,0,sigma) # residuals

MitoSize <- Y0*CellSize^alpha + eps
df <- data.frame(CellSize,MitoSize)
ggplot(df,aes(CellSize,MitoSize)) + geom_point()
```

## Part 1 

__(a)__ Use nonlinear least squares (nls) to estimate the parameter $\alpha$ from the simulated data.

```r
m <- nls(MitoSize ~ 100*CellSize^alph,data=df,start=list(alph=1))
```

```r
coef(m) > 0
```

```r
abs(-1)
```

```r
summary(m)
```

##### Define the __error__ as $e=|\hat{\alpha}-\alpha|$ where $\hat{\alpha}$ is the estimate and $\alpha$ is the true value (which is known because this is simualted data). 

__(b)__ Write a function to execute the simulation and nls fitting, taking in the true model parameters $\alpha, y_0, \sigma$ and the sample size $N$. Plot the error in $\alpha$ estimation, $e$, as a function of the magnitude of the noise, $\sigma$, for a range of $\sigma=(100,2000)$. 

```r
errF <- function(alpha, y0, N, sigma) {
    CellSize <- runif(N,3,10) # independent variable
    eps <- rnorm(N,0,sigma) # residuals
    MitoSize <- y0*CellSize^alpha + eps
    df <- data.frame(CellSize,MitoSize)
    m <- nls(MitoSize ~ 100*CellSize^alph,data=df,start=list(alph=1))
    abs(alpha-coef(m))/alpha
}
```

```r
N <- 30
alpha <- 2.3
Y0 <- 100
# for loop over sigma from 100 to 2000
loop.len <- 1000
loop.vec <- seq(from = 100, to = 2000, length.out = loop.len)
est.vec <- seq(from = 1, to = loop.len, length.out = loop.len)
for (i in 1:loop.len) {
    est.vec[i] <- errF(alpha, Y0, N, loop.vec[i])
}
```

```r
dfp1 <- data.frame(loop.vec, est.vec)
ggplot(dfp1,aes(loop.vec,est.vec)) + geom_point()
```

## Part 2

__(a)__ Transform the simulated data to $lY = \mbox{log}(y)$ and $lX = \mbox{log}(x)$. Use linear regression model fit (lm) to estimate the parameter $\alpha$ from the simulated data.

```r
# plot settings
options(repr.plot.width=4, repr.plot.height=3)

# Simulate power-law data

N <- 30
alpha <- 2.3
Y0 <- 100
sigma <- 500

CellSize <- runif(N,3,10) # independent variable
eps <- rnorm(N,0,sigma) # residuals

MitoSize <- Y0*CellSize^alpha + eps
lY <- log(MitoSize)
lX <- log(CellSize)

plot(CellSize,MitoSize)
plot(lX,lY)
```

```r
df2 <- data.frame(lX,lY)
m <- lm(lY ~ lX,data=df2)
coef(m)[2]
```

__(b)__ Write a function that executes the simulation, log-transforms the data, and performs linear regression fitting. Plot the error in $\alpha$ estimation, $e$, as a function of the magnitude of the noise, $\sigma$, for a range of $\sigma=(100,2000)$. 

Note you may need additional code to handle negative measurements.

__(c)__ Overplot the error in $\alpha$ from nls and the error from lm (as functions of $\sigma$). 

```r
# b)
errLF <- function(alpha, y0, N, sigma) {
    CellSize <- runif(N,3,10) # independent variable
    eps <- rnorm(N,0,sigma) # residuals
    MitoSize <- y0*CellSize^alpha + eps
    flag <- MitoSize > 0
    lX <- log(CellSize[flag])
    lY <- log(MitoSize[flag])
    df2 <- data.frame(lX,lY)
    m <- lm(lY ~ lX,data=df2)
    abs(alpha-coef(m)[2])/alpha
}
```

```r
N <- 30
alpha <- 2.3
Y0 <- 100
# for loop over sigma from 100 to 2000
estL.vec <- seq(from = 1, to = loop.len, length.out = loop.len)
for (i in 1:loop.len) {
    estL.vec[i] <- errLF(alpha, Y0, N, loop.vec[i])
}
```

```r
dfp1a <- data.frame("s"=loop.vec,"RelError"=est.vec); dfp1a["RegMethod"] <- "nls"
dfp1b <- data.frame("s"=loop.vec,"RelError"=estL.vec); dfp1b["RegMethod"] <- "lm"
dfp2 <- full_join(dfp1a, dfp1b, by = c("s", "RelError", "RegMethod"))
head(dfp2)
dim(dfp2)
```

```r
ggplot(dfp2,aes(x=s,y=RelError,colour=RegMethod)) + # basic graphical object
  geom_point() + # first layer
  scale_colour_manual(values=list("blue","green")) # modifying colours
```

## Part 3

In the above simulations, the independent variable $X$ was chosen from a uniform distribution.

__(a)__ Find a distribution of $X$ data that maximizes the difference between error from nls and error from lm.

__(b)__ Find a distribution of $X$ data that minimizes the difference between the error from nls and error from lm.

```r
# a)
errFMax <- function(alpha, y0, N, sigma) {
    CellSize <- rlnorm(N,0,1) # rexp(N,0.5)-0.25, rlnorm(N,0,1)-1.5
    eps <- rnorm(N,0,sigma) # residuals
    MitoSize <- y0*CellSize^alpha + eps
    df <- data.frame(CellSize,MitoSize)
    m <- nls(MitoSize ~ y0*CellSize^alph,data=df,start=list(alph=1))
    abs(alpha-coef(m))/alpha
}
```

```r
N <- 30
alpha <- 2.3
Y0 <- 100
# for loop over sigma from 100 to 2000
loop.len <- 1000
loop.vec <- seq(from = 100, to = 2000, length.out = loop.len)
est.vec <- seq(from = 1, to = loop.len, length.out = loop.len)
for (i in 1:loop.len) {
    est.vec[i] <- errFMax(alpha, Y0, N, loop.vec[i])
}
```

```r
dfp1m <- data.frame(loop.vec, est.vec)
ggplot(dfp1m,aes(loop.vec,est.vec)) + geom_point()
```

```r
# b)

errFMin <- function(alpha, y0, N, sigma) {
    CellSize <- rnorm(N,100,5) # also: rgamma(N,100)
    eps <- rnorm(N,0,sigma) # residuals
    MitoSize <- y0*CellSize^alpha + eps
    df <- data.frame(CellSize,MitoSize)
    m <- nls(MitoSize ~ y0*CellSize^alph,data=df,start=list(alph=1))
    abs(alpha-coef(m))/alpha
}

N <- 30
alpha <- 2.3
Y0 <- 100
# for loop over sigma from 100 to 2000
loop.len <- 1000
loop.vec <- seq(from = 100, to = 2000, length.out = loop.len)
est.vec <- seq(from = 1, to = loop.len, length.out = loop.len)
for (i in 1:loop.len) {
    est.vec[i] <- errFMin(alpha, Y0, N, loop.vec[i])
}

dfp2m <- data.frame(loop.vec, est.vec)
ggplot(dfp2m,aes(loop.vec,est.vec)) + geom_point()
```
